```python
from dotenv import load_dotenv
import os

load_dotenv()
NEON_CONN_STRING = os.getenv("NEON_CONN_STRING")

# quick sanity check - just confirms it loaded, without printing the real password
print("Connection string loaded:", NEON_CONN_STRING is not None)
```

```python
import subprocess
import os
from datetime import datetime

os.makedirs("backups", exist_ok=True)

timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
backup_file = f"smakjit_backup_{timestamp}.dump"

result = subprocess.run(
    [
        "pg_dump",
        NEON_CONN_STRING,
        "--format=custom",
        f"--file={backup_file}"
    ],
    capture_output=True,
    text=True
)

print("Return code:", result.returncode)
print("STDOUT:", result.stdout)
print("STDERR:", result.stderr)
print("Backup file:", backup_file)
```
